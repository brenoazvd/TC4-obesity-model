```python
cross_tab = pd.crosstab(df['family_history'], df['Obesity'], normalize='index')

ax = cross_tab.plot(
    kind='bar', 
    stacked=True, 
    figsize=(12, 7), 
    colormap='RdYlBu_r', 
    edgecolor='black',   
    alpha=0.85
)


plt.title('Impacto do Histrico Familiar na Distribuio da Obesidade', fontsize=16)
plt.xlabel('Possui Histrico Familiar?', fontsize=12)
plt.ylabel('Proporo de Casos', fontsize=12)
plt.xticks(rotation=0) 
plt.legend(title='Classificao', bbox_to_anchor=(1.05, 1), loc='upper left')


for n, x in enumerate([*cross_tab.index.values]):
    for (proportion, y_loc) in zip(cross_tab.loc[x], cross_tab.loc[x].cumsum()):
        
        if proportion > 0.01: 
            plt.text(x=n,
                     y=(y_loc - proportion) + (proportion / 2),
                     s=f'{proportion:.1%}', 
                     color="black",
                     fontsize=9,
                     ha="center", 
                     va="center")

plt.tight_layout()
plt.show()
```

```python

# 1. Definir a ordem lgica (Crucial para ver a progresso)
ordem_categorias = [
    'Insufficient_Weight', 'Normal_Weight', 'Overweight_Level_I', 
    'Overweight_Level_II', 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III'
]

# Configurar a rea de plotagem (2 grficos lado a lado)
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# --- GRFICO 1: Consumo de Alimentos Calricos (FAVC) ---
# Vamos calcular a % de pessoas que responderam "yes" para FAVC em cada grupo
favc_percent = pd.crosstab(df['Obesity'], df['FAVC'], normalize='index')['yes'] * 100

# Reordenar e plotar
favc_percent = favc_percent.reindex(ordem_categorias)
bars = favc_percent.plot(kind='bar', color='#ff6f69', ax=ax[0], edgecolor='black')

ax[0].set_title('Percentual de Consumo de Alimentos Calricos (FAVC = Yes)', fontsize=14)
ax[0].set_ylabel('% de Pessoas', fontsize=12)
ax[0].set_xlabel('')
ax[0].grid(axis='y', linestyle='--', alpha=0.5)

# Adicionar rtulos nas barras
for container in ax[0].containers:
    ax[0].bar_label(container, fmt='%.1f%%', padding=3)

# --- GRFICO 2: Consumo de Vegetais (FCVC) ---
# Boxplot para ver a mediana e a disperso dos dados (1 = pouco, 3 = muito)
sns.boxplot(
    data=df, 
    x='Obesity', 
    y='FCVC', 
    order=ordem_categorias, 
    ax=ax[1], 
    palette='Greens',
    hue='Obesity',
    medianprops={"color": "gold", "linewidth": 1.5}
)

ax[1].set_title('Distribuio do Consumo de Vegetais (FCVC) com Mediana Destacada', fontsize=14)
ax[1].set_ylabel('Frequncia de Consumo (1 a 3)', fontsize=12)
ax[1].set_xlabel('')
ax[1].tick_params(axis='x', rotation=45) # Rotacionar rtulos para leitura

plt.tight_layout()
plt.show()
```

```python
Nesta visualizao mostra o quanto os grupos comem de alimentos caloricos e vegetais:

* Como observamos o grupo de obesidade nivel tres tem o maior indice de consumo de alimentos caloricos `FAVC` - 99.7% 
comparando com o grupo de peso normal cai para 72,5%  indicando que pessoas com peso normal tendem a consumir menos alimentos caloricos.

O grupo obesidade nivel trs tambm apresenta uma frequencia de consumo de vegetais grande, ento indicam que mantem uma dieta rica em hipercalricos e rica em vegetais.
```

```python
#FAF -> FISICAL ACTIVITY FREQUENCY
#TUE -> TIME USING TECHNOLOGY

df_agr_faf_tue=df.groupby('Obesity')[['FAF', 'TUE']].mean().sort_values(by='FAF')
df_agr_faf_tue.reset_index(inplace=True)
df_agr_faf_tue['FAF']=df_agr_faf_tue['FAF'].astype('float')
df_agr_faf_tue['TUE']=df_agr_faf_tue['TUE'].astype('float')
df_agr_faf_tue.sort_values(by='TUE',ascending=False)

```

Notamos que obesidade nivel trs tem o menor indice de prtica de atividade fsica disparado de 0.63 enquanto o minimo dos outros grupos  de 0.96

*Pessoas com peso normal se exercitam o dobro do que o nivel 3 de obsdd.

entendemos que para casos graves (nivel 3), essa disparidade em pratica de atividade fisica mostra-se como determinante

Nessa tabela vemos tambm que o uso de tecnologia no condiz com o tipo de obsdd, por conta de o grupo de peso baixo ficar cerca de 84% do tempo enquanto no h uma correlao direta com nvel de obsdd


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o grfico
plt.figure(figsize=(10, 6))

# Plotar os pontos (Scatter)
# Eixo X = Atividade Fsica (FAF), Eixo Y = Tecnologia (TUE)
sns.scatterplot(
    data=df_agr_faf_tue, 
    x='FAF', 
    y='TUE', 
    hue='Obesity', # Colore cada ponto conforme a categoria
    s=200,         # Tamanho dos pontos
    palette='RdYlBu_r', # Cores variadas
    edgecolor='black'
)

# Adicionar rtulos em cada ponto para saber quem  quem
# Iteramos sobre as linhas do seu df agrupado
for i in range(df_agr_faf_tue.shape[0]):
    plt.text(
        x=df_agr_faf_tue.FAF[i]+0.02, # Posio X (+ um pouquinho p/ direita)
        y=df_agr_faf_tue.TUE[i],      # Posio Y
        s=df_agr_faf_tue.Obesity[i],  # O texto (nome da categoria)
        fontdict=dict(color='black', size=10),
        bbox=dict(facecolor='white', alpha=0.5, edgecolor='none') # Fundo branco p/ leitura
    )

# Melhorias visuais
plt.title('Mapa de Hbitos: Sedentarismo vs. Uso de Tecnologia', fontsize=14)
plt.xlabel('Frequncia de Atividade Fsica (Mdia)', fontsize=12)
plt.ylabel('Tempo de Uso de Tecnologia (Mdia)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Remover a legenda automtica (j colocamos os nomes nos pontos)
plt.legend([],[], frameon=False)

plt.tight_layout()
plt.show()
```

Entendemos que para combater a obsdd severa a atividade fisica, tem mais importancia do que o uso de telas.

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Tabela Cruzada: Transporte vs Obesidade
# Queremos saber: "Dentre quem usa Carro, como  a obesidade?"
cross_transporte = pd.crosstab(df['MTRANS'], df['Obesity'], normalize='index')

# 2. Reordenar as colunas de obesidade (para o degrad fazer sentido)
ordem_categorias = [
    'Insufficient_Weight', 'Normal_Weight', 'Overweight_Level_I', 
    'Overweight_Level_II', 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III'
]
colunas_existentes = [col for col in ordem_categorias if col in cross_transporte.columns]
cross_transporte = cross_transporte[colunas_existentes]

# 3. Filtrar Transportes irrelevantes (Opcional)
# s vezes aparecem 'Bike' ou 'Motorbike' com pouqussimos dados que sujam o grfico.
# Vamos focar nos principais: Automobile, Public_Transportation, Walking
# Se quiser ver todos, comente a linha abaixo.
principais_meios = ['Automobile', 'Public_Transportation', 'Walking']
cross_transporte = cross_transporte.loc[cross_transporte.index.isin(principais_meios)]

# 4. Plotar
ax = cross_transporte.plot(
    kind='bar', 
    stacked=True, 
    figsize=(12, 7), 
    colormap='RdYlBu_r', 
    edgecolor='black',
    alpha=0.9
)

plt.title('Influncia do Meio de Transporte no Grau de Obesidade', fontsize=16)
plt.ylabel('Proporo de Casos', fontsize=12)
plt.xlabel('Meio de Transporte Utilizado', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Classificao', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionar rtulos (apenas para fatias relevantes > 5%)
for n, x in enumerate([*cross_transporte.index.values]):
    for (proportion, y_loc) in zip(cross_transporte.loc[x], cross_transporte.loc[x].cumsum()):
        if proportion > 0.01: 
            plt.text(x=n,
                     y=(y_loc - proportion) + (proportion / 2),
                     s=f'{proportion:.1%}', 
                     color="black",
                     fontsize=9,
                     fontweight="bold",
                     ha="center", 
                     va="center")

plt.tight_layout()
plt.show()
```
